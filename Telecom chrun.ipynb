```python
import pandas as pd
import matplotlib.pyplot as plt

# Compute percentages per group
gb = (
    df
      .groupby(['SeniorCitizen', 'Churn'])
      .size()
      .div(df.groupby('SeniorCitizen').size(), level=0)
      .mul(100)
      .round(1)
      .rename('percent')
      .reset_index()
)

# Pivot for stacked plotting
pivot = gb.pivot(index='SeniorCitizen', columns='Churn', values='percent')

# Plot with custom colors: blue then orange
ax = pivot.plot(
    kind='bar',
    stacked=True,
    figsize=(6, 5),
    color=['orange', 'blue']
)

# Add centered percentage labels in each segment
for c in ax.containers:
    labels = [f'{h:.1f}%' if h > 0 else '' for h in c.datavalues]
    ax.bar_label(c, labels=labels, label_type='center')

ax.set_xlabel('SeniorCitizen')
ax.set_ylabel('Percent within group')
ax.set_title('Churn by SeniorCitizen (percentage)')
ax.legend(title='Churn')
plt.xticks(rotation=0)
plt.legend(title = "Churn" , bbox_to_anchor = (0.9,0.9))
plt.show()
```

```python
import pandas as pd

# List of columns to plot (excluding 'customerID' if it's unique for each row)
cols = ['gender', 'SeniorCitizen', 'Partner', 'Dependents',
        'PhoneService', 'MultipleLines', 'InternetService',
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up the subplot grid
n_cols = 3
n_rows = -(-len(cols) // n_cols)  # Ceiling division

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))
axes = axes.flatten()

# Create countplots for each column
for i, col in enumerate(cols):
    sns.countplot(data=df, x=col, ax=axes[i], palette='Set2', hue = df["Churn"])
    axes[i].set_title(f"Count of {col}")
    axes[i].tick_params(axis='x', rotation=45)

# Remove unused axes (if any)
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

The Majority of customer who do not churn tend to have like phone service internet service (particular DSL) and online security enabled
online backup tech support and streaming tv churn rate are noticeably higher when these services are not or are unavailable
